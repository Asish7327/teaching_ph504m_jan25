#### Writing a code to fit a straight line y = mx + c

##### The equation derived in the class for least square approximation
### $ c = \frac{\sum{x_i^2} \sum{y_i} - \sum{x_i} \sum{x_i y_i}}{n\sum{x_i^2} - (\sum{x_i})^2} $
### $ m = \frac{n\sum{x_i y_i} - \sum{x_i} \sum{y_i}}{n\sum{x_i^2} - (\sum{x_i})^2} $

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Let's creat a data (Teacher mode)
# x values -- the independent variable
x = np.arange(-4, 4.5, 0.5)
# y values -- let's generate from y = 2 x + 3 line
# but I can't keep all the points on the same line so I will be randomly shifting the data
np.random.seed(42)
y = (2*x + 3) + np.random.normal(0, 1, len(x))
```

```python
# check out the data by plotting
plt.scatter(x, y, s = 40, color = 'black', facecolor = 'cyan')
plt.xlabel('X')
plt.ylabel('Y')
```

### code to fit a straight line with least square

```python
def least_square_fit(x, y):
    n = len(x)
    denominator =  n* np.sum(x**2) - (np.sum(x))**2

    slope = (n * np.sum(x*y) - np.sum(x) * np.sum(y)) / denominator
    intercept = (np.sum(x**2) * np.sum(y) - np.sum(x) * np.sum(x*y) ) / denominator

    return slope, intercept
    
```

```python
m, c =  least_square_fit(x, y)
```

```python
print(m, c)
```

```python
# check out the data by plotting
plt.scatter(x, y, s = 40, color = 'black', facecolor = 'cyan')
ynew = m*x + c
plt.plot(x, ynew, linestyle = '--', linewidth = 2, color = 'red', label = 'LS fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
