### Answer 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate sum of squared residuals
def calculate_f(m, c):
    # hardcoding the equation here
    x = np.linspace(0, 10, 50)  # Generate some x values
    y_actual = 4*x + 5  # True y values
    
    y_pred = m*x + c
    return np.sum((y_actual - y_pred) ** 2)


# 1. Minima with respect to slope (m)
m_values = np.arange(1, 7.1, 0.1)

f_m =[]
for m in m_values:
    f = calculate_f(m, 5)
    f_m.append(f)

# minima
index = np.argmin(f_m)
print("Minima of f is at m =", m_values[index], ' and minimum value of f is:', f_m[index])

# Plotting
plt.plot(m_values, f_m, label="f vs. slope (m)", color="blue")
plt.axvline(x=4, linestyle="--", color="red", label="True m = 4")
plt.axhline(y =f_m[index], linestyle="--", color="gray" )
plt.xlabel("Slope (m)")
plt.ylabel("Sum of Squared Residuals (f)")
plt.legend()
plt.title("Minimization of f with respect to slope (m)")
plt.show()

# 2. Minima with respect to intercept (c)
c_values = np.arange(2, 8.1, 0.1)

f_c =[]
for c in c_values:
    f = calculate_f(4, c)
    f_c.append(f)

# minima
index = np.argmin(f_c)
print("Minima of f is at c =", c_values[index], ' and minimum value of f is:', f_c[index])

#plotting
plt.plot(c_values, f_c, label="f vs. intercept (c)", color="green")
plt.axvline(x=5, linestyle="--", color="red", label="True c = 5")
plt.axhline(y =f_c[index], linestyle="--", color="gray" )
plt.xlabel("Intercept (c)")
plt.ylabel("Sum of Squared Residuals (f)")
plt.legend()
plt.title("Minimization of f with respect to intercept (c)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants

# Lambertian reflection model
def brightness(d, theta):
    """
    d in meters
    theta in degrees
    """
    I0 = 1  # Assume intrinsic brightness in arbitrary units
    R = 6.37e6  # Earth's radius (meters)
    return I0 * (R**2 / d**2) * np.cos(np.radians(theta))


d_values = np.arange(1.0, 101.0, 1)  # Distances in km (1 AU, 10 AU, 40 AU, 100 AU)
d_values *= 1.5e11  # Convert km to meters
angles = [0, 32, 60]  # Phase angles in degrees

# Calculate brightness for different angles and plot
for theta in angles:
    intensities = []
    for d in d_values:
        intensities.append(brightness(d, theta))

    plt.plot(d_values / 1.5e11, intensities, label="Angle = " + str(theta) + " degrees")

plt.axvline(x=6e12/1.5e11, linestyle="--", color="red")
plt.scatter(x= [6e12/1.5e11], y= [brightness(6e12, 32)], color = 'blue', marker = 'o', label = "Earth's Voyager brightness")

plt.xlabel("Distance from Earth in AU")
plt.ylabel("Relative Brightness")
plt.yscale("log")
plt.legend()
plt.title("Brightness of Earth as Seen from Voyager 1")
plt.show()
```
